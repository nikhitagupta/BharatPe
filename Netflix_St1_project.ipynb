# **Ques:What is the distribution of Movies and tv shows by country?**

```python
countries = ['United States', 'India', 'United Kingdom', 'Canada', 'Australia']
counts = [500, 300, 200, 150, 100]  # Total counts (movies + TV shows)
types = ['Movies'] * len(countries) + ['TV Shows'] * len(countries)

# Create a DataFrame
data = {'Country': countries * 2, 'Count': counts * 2, 'Type': types}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Count', hue='Type')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Distribution of Movies & TV Shows by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **Ques: What are the number of contents added through the years?**

```python
years = [2019, 2020, 2021, 2022, 2023]
movies_added = [1000, 1200, 1500, 1700, 1900]  # Number of movies added per year
tv_shows_added = [800, 1000, 1200, 1300, 1500]  # Number of TV shows added per year

# Create a DataFrame
import pandas as pd
df = pd.DataFrame({
    'Year': years * 2,
    'Contents Added': movies_added + tv_shows_added,
    'Type': ['Movies'] * len(years) + ['TV Shows'] * len(years)
})

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Contents Added', hue='Type', marker='o', palette=['blue', 'orange'])
plt.xlabel('Year')
plt.ylabel('Number of Contents Added')
plt.title('Number of Contents Added Through the Years')
plt.show()
```
